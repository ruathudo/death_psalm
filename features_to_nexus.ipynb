```python
import pandas as pd

featurelist = pd.read_csv("data/features.csv", skiprows=4, header=1)
```

```python
source_columns = list(featurelist.columns[6::2])
short_names = [s.split(' ')[0].replace('ö','o').replace('ä','a').replace("'",'')+'-' for s in source_columns]
short_names
```

```python
sources = featurelist[ source_columns ]
sources.fillna('?', inplace=True)
sources.replace('CH','?', inplace=True)
sources = sources.T
sources.index = short_names
taxlabels = " ".join([s.replace('-','') for s in sources.index])
```

```python
data = sources.to_string(header=False).replace(' ', '').replace('-',' ')
```

```python
nexus = """
#NEXUS

BEGIN TAXA;
    DIMENSIONS NTAX={ntax};
    TAXLABELS {taxlabels};
END;

BEGIN CHARACTERS;
    DIMENSIONS NCHAR={nchar};
    FORMAT DATATYPE=standard symbols="0 1" MISSING=?;
    
    
MATRIX
  {data}
;
END;
""".format(ntax=sources.shape[0], nchar=sources.shape[1], taxlabels=taxlabels, data=data)
print(nexus)
f = open('data/lalli.nex', 'w')
f.write(nexus)
f.close()
```
